```python
dictPic = {}
dictPic["class"]    = "corn"
dictPic["filename"] =  "multi_or.png"
dictPic["annotations"] = []

labels = []
with open("multi_or.txt") as file:
    for line in file.readlines():
        dictLine = {}
        for index,each in enumerate(line.rsplit()):
            if index == 1:
                dictLine["x"] = (int((float(each)*5220)))
            elif index == 2:
                dictLine["y"] = (int((float(each)*5672)))
            elif index == 3:
                dictLine["width"] = (int((float(each)*5220)))
            elif index == 4:
                dictLine["height"] = (int((float(each)*5672)))
            else:
                dictLine["class"] = "rect"
            print(f"printing char {index}  ")
        dictPic["annotations"].append(dictLine)
```
